```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\drive-download-20250124T081253Z-001\Admission_Predict.csv")
```

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['GRE_Score'], y=df['Chance_of_Admit'], alpha=0.6, color='b')
plt.xlabel('GRE Score')
plt.ylabel('Chance of Admit')
plt.title('Scatter Plot: GRE Score vs Chance of Admit')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x=df['University_Rating'], y=df['Chance_of_Admit'], errorbar=None, palette='coolwarm', hue=df['University_Rating'], dodge=False)
plt.xlabel('University Rating')
plt.ylabel('Average Chance of Admit')
plt.title('Bar Plot: University Rating vs Chance of Admit')
plt.legend([],[], frameon=False)  # Hide legend
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['CGPA'], color='green')
plt.ylabel('CGPA')
plt.title('Box Plot: Distribution of CGPA')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['Research'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightblue', 'lightcoral'], labels=['No Research', 'Has Research'])
plt.title('Pie Chart: Research vs No Research')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=df['GRE_Score'], y=df['Chance_of_Admit'], marker='o', color='purple')
plt.xlabel('GRE Score')
plt.ylabel('Chance of Admit')
plt.title('Line Chart: GRE Score vs Chance of Admit')
plt.show()
```
